## **2025-11-03-Create-Small-Lib-For-Chroms**

Create a small library for the chromatogram analysis. Based on what is identified in 1000pg replicate #2. 

```python
import polars as pl
```

```python
lib_orig = pl.scan_csv("../PeptDeep-NoMods-In-Silico-Library-Generation/2025-06-10-in-silico-lib-no-mods_fix_mods_filtered_6Frags_decoys_filtered.tsv", separator='\t', infer_schema_length=100000000)
```

```python
lib_refined = pl.read_csv('../PeptDeep-NoMods-Refined-MasterLib-Analysis/osw/pqp_libraries/HeLa02DDM_500pg_5x3_PyDIA_1_S1-C10_1_1642_osw_4_6Frags.tsv', separator='\t', infer_schema_length=1000000)
```

```python
lib_refined = lib_refined.with_columns((pl.col('ModifiedPeptideSequence') + pl.col('PrecursorCharge').cast(pl.Utf8)).alias("Precursor"))
```

```python
precs_to_filter_to = lib_refined['Precursor'].unique().to_list()
```

```python
lib_orig_filtered = (lib_orig
 .with_columns((pl.col('ModifiedPeptideSequence') + pl.col('PrecursorCharge').cast(pl.Utf8)).alias("Precursor"))
 .filter(pl.col('Precursor').is_in(precs_to_filter_to))
 .collect()
                     
)
         
```

```python
lib_orig_filtered = lib_orig_filtered.with_columns((pl.col("FragmentType") + pl.col("FragmentSeriesNumber").cast(pl.Utf8) + "^" + pl.col("ProductCharge").cast(pl.Utf8)).alias("Annotation"))
```
