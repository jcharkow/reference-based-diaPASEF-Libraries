```python
ng25_experiment = OpenSwathXICParquetLoader(rsltsFile="../../results/Dilutions-PeptDeep-NoMods-Lib/osw/90min-SP-30cm-2um-K562-100nL-25ng_DIA_Slot1-5_1_550_3-7-2021/pyprophet_LDA_XGBoost/90min-SP-30cm-2um-K562-100nL-25ng_DIA_Slot1-5_1_550_3-7-2021.oswpq",
                                           dataFiles="../../results/Dilutions-PeptDeep-NoMods-Lib/osw/90min-SP-30cm-2um-K562-100nL-25ng_DIA_Slot1-5_1_550_3-7-2021/oswOut/90min-SP-30cm-2um-K562-100nL-25ng_DIA_Slot1-5_1_550_3-7-2021.sqMass_xic.parquet")
```

----

##### **Load 25ng Chromatogram**

```python
ng_25_chromatograms = ng25_experiment.loadTransitionGroupsDf(prec, chg)

ng_25_chromatograms["intensity"] = (
    ng_25_chromatograms
    .groupby("annotation")["intensity"]
    .transform(lambda x: savgol_filter(x, window_length=11, polyorder=3))
)


ng_25_chromatograms = ng_25_chromatograms[(~ng_25_chromatograms['annotation'].str.contains('Precursor'))]
```

```python
ng_25_chromatograms.to_pickle("ng_25_chromatogram.pkl")
```
