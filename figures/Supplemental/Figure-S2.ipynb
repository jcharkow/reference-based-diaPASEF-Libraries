```python
graph_points = np.load('2021-01-25-graphPoints.npy') # density of library experimental library entries
```

```python
def plot_sampling_scheme(ax):
    colors = ['tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:olive', 'tab:gray', 'tab:cyan', 'chocolate', 'cornflowerblue', 'indigo', 'springgreen']
    labels = [ "Injection: #{}".format(i + 1) for i in range(len(colors))]
    windows_patches = []

    legend_elements = []
    for c, l in zip(colors, labels):
        legend_elements.append(Patch(facecolor=c, label=l))

    for c, w in zip(colors, windows):
        windows_patches.append(w.getPatches(alpha=0.2, color=c, lw=1))

    ax.scatter(graph_points[0], graph_points[1], c=graph_points[2], s=1, marker='.', cmap='Blues')
    ax.set_xlabel("m/z", fontsize='large')
    ax.set_ylabel("Ion Mobility (1/k0)", fontsize='large')

    for exp in windows_patches:
        for patch in exp:
            ax.add_patch(patch)


    ax.legend(handles=legend_elements, bbox_to_anchor=(1,1), fontsize='xx-small')
```

```python
fig, ax = plt.subplots(1,1, figsize=(8, 5), dpi=300)
plot_sampling_scheme(ax)
plt.savefig("Figure-S2.png", dpi=300, bbox_inches='tight')
```
