This is a helper script which fetches the chromatogram data for the dilutions. 

```python
from massdash.loaders import OpenSwathXICParquetLoader
from massdash.loaders import SqMassLoader
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import massdash
import sys
```

```python
print("MassDash Version", massdash.__version__)
print("Pandas Version", pd.__version__)
print("Python Version", sys.version)
```

```python
prec = 'YNPENLATLER'
chg = 2
```

```python
pg500_orig = SqMassLoader(rsltsFile=["../../results/SCP-PeptDeep-NoMods-Analysis/osw_xtraChroms/HeLa02DDM_500pg_5x3_PyDIA_1_S1-C10_1_1642/HeLa02DDM_500pg_5x3_PyDIA_1_S1-C10_1_1642.osw"],
                      dataFiles="../../results/SCP-PeptDeep-NoMods-Analysis/osw_xtraChroms/HeLa02DDM_500pg_5x3_PyDIA_1_S1-C10_1_1642/HeLa02DDM_500pg_5x3_PyDIA_1_S1-C10_1_1642.sqMass")
```

```python
pg100_refined = SqMassLoader(rsltsFile=["../../results/SCP-PeptDeep-NoMods-Refined-Analysis/osw_xtraChroms/HeLa02DDM_100pg_5x3_PyDIA_1_S1-C4_1_1630/HeLa02DDM_1000pg_5x3_PyDIA_2_S1-D2_1_1649_lib/HeLa02DDM_100pg_5x3_PyDIA_1_S1-C4_1_1630.osw"],
                      dataFiles="../../results/SCP-PeptDeep-NoMods-Refined-Analysis/osw_xtraChroms/HeLa02DDM_100pg_5x3_PyDIA_1_S1-C4_1_1630/HeLa02DDM_1000pg_5x3_PyDIA_2_S1-D2_1_1649_lib/HeLa02DDM_100pg_5x3_PyDIA_1_S1-C4_1_1630.sqMass")
```

----

##### **Load 100pg Chromatogram**

```python
pg100_chromatograms = pg100_refined.loadTransitionGroupsDf(prec, chg)


pg100_chromatograms["intensity"] = (
    pg100_chromatograms
    .groupby("annotation")["intensity"]
    .transform(lambda x: savgol_filter(x, window_length=11, polyorder=3))
    .clip(0)
)
pg100_chromatograms = pg100_chromatograms[(~pg100_chromatograms['annotation'].str.contains('Precursor'))]
```

```python
pg100_chromatograms.to_pickle("pg_100_chromatogram.pkl")
```

##### **Load 1000pg Chromatogram**

```python
pg_500_chromatograms = pg500_orig.loadTransitionGroupsDf(prec, chg)


pg_500_chromatograms["intensity"] = (
    pg_500_chromatograms
    .groupby("annotation")["intensity"]
    .transform(lambda x: savgol_filter(x, window_length=11, polyorder=3))
    .clip(0)
)
pg_500_chromatograms = pg_500_chromatograms[(~pg_500_chromatograms['annotation'].str.contains('Precursor'))]
```

```python
pg_500_chromatograms.to_pickle("pg_500_chromatogram.pkl")
```
